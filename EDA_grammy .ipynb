```python
import pandas as pd 
import numpy 
import matplotlib.pyplot as plt
import psycopg2
import json 
```

```python
def connect_postgres():
    with open('C:/Users/SEBASTIAN/OneDrive/Escritorio/Workshop02/db_config.json') as f:
        dbfile = json.load(f)
    connection = psycopg2.connect(
        database=dbfile["database"],
        user=dbfile["user"],
        password=dbfile["password"],
        host="localhost",
        port=5432
    )
    print("Database connection ok")
    return connection
    
```

```python
connection = connect_postgres()
cursor = connection.cursor()
query1 = cursor.execute("""SELECT * FROM grammy""")
connection.commit()
resultados = cursor.fetchall()
df = pd.DataFrame(resultados)
df.columns = ['year','title','published_at','updated_at','category','nominee','artist','workers','img','winner']
```

```python
non_null_count = df.count().sum()
null_count = df.isnull().sum().sum()
data = {'Valores Presentes': non_null_count, 'Nulos': null_count}
data_series = pd.Series(data)

plt.figure(figsize=(8, 8))
plt.pie(data_series, labels=data_series.index, autopct='%1.1f%%', startangle=140)
plt.title('Valores No Nulos vs. Nulos en el DataFrame')
plt.show()
```
