<span style="font-family: Arial; font-size: 20px; font-weight: bold;">Tabla canciones mas repetidas.</span>

<span style="font-family: Arial; font-size: 15px; font-weight: bold;">En la tabla siguiente, se evidencia la presencia de canciones con el mismo nombre y artista, algunas de las cuales se repiten hasta 150 veces. Debido a esta duplicacin, se proceder a agrupar las canciones segn su nombre y artista, y se seleccionarn aquellas con la mayor popularidad entre las duplicadas</span>

```python
duplicadas = spotify_df[spotify_df.duplicated(subset=['artists', 'track_name'], keep=False)]
conteo_duplicados = duplicadas[['artists', 'track_name']].value_counts().reset_index(name='Nmero de Repeticiones')
conteo_duplicados = conteo_duplicados.sort_values(by='Nmero de Repeticiones', ascending=False)
top_10_duplicados = conteo_duplicados.head(10)
tabla_html = top_10_duplicados.to_html(index=False)
display(HTML(tabla_html))
```

<span style="font-family: Arial; font-size: 20px; font-weight: bold;">Tabla de Canciones Populares Repetidas</span>


```python
duplicadas = spotify_df[spotify_df.duplicated(subset=['artists', 'track_name'], keep=False)]
duplicadas = duplicadas.sort_values(by='danceability', ascending=False)
top_10_populares = duplicadas.head(10)
tabla_html = top_10_populares[['artists', 'track_name', 'danceability']].to_html(index=False)
display(HTML(tabla_html))
conteo_populares = top_10_populares.groupby(['artists', 'track_name'])['danceability'].count().reset_index(name='Nmero de Repeticiones')
tabla_html_conteo = conteo_populares[['artists', 'track_name', 'Nmero de Repeticiones']].to_html(index=False)
display(HTML(tabla_html_conteo))
```

<span style="font-family: Arial; font-size: 15px; font-weight: bold;">Despus de completar este proceso, se puede notar una clara diferencia cuando consultamos las canciones ms bailables, estas aparecen solo una vez, a diferencia de la tabla anterior.</span>

```python
spotify_df = spotify_df.sort_values(by='danceability', ascending=False)
spotify_df = spotify_df.drop_duplicates(subset=['artists', 'track_name'], keep='first')
tabla_html = spotify_df[['artists', 'track_name', 'danceability']].head(20).to_html(index=False)
display(HTML(tabla_html))
```
